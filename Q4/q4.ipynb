```python
import csv
def main() :
    f = open('q4.csv')
    data = csv.reader(f)
    header = next(data)
    bs1 =''
    bs1u =0
    lbs1 = ''
    lbs1u = 99999999
    bs2 = ''
    bs2u = 0
    lbs2 = ''
    lbs2u = 99999999
    bs3 = ''
    bs3u = 0
    lbs3 = ''
    lbs3u = 99999999
    bs4 = ''
    bs4u = 0
    lbs4 = ''
    lbs4u = 99999999
    user = 0
    for row in data:
        if row[1] == "1호선":
            user = int(row[4])+int(row[5])
            if user > bs1u :
                bs1 = row[3]
                bs1u = user
            if user < lbs1u :
                lbs1 = row[3]
                lbs1u = user
        
        if row[1] == "2호선":
            user = int(row[4])+int(row[5])
            if user > bs2u :
                bs2 = row[3]
                bs2u = user
            if user < lbs2u :
                lbs2 = row[3]
                lbs2u = user
        
        if row[1] == "3호선":
            user = int(row[4])+int(row[5])
            if user > bs3u :
                bs3 = row[3]
                bs3u = user
            if user < lbs3u :
                lbs3 = row[3]
                lbs3u = user
        
        if row[1] == "4호선":
            user = int(row[4])+int(row[5])
            if user > bs4u :
                bs4 = row[3]
                bs4u = user
            if user < lbs4u :
                lbs4 = row[3]
                lbs4u = user
                
        
            
          
            
    
    print("***Subway Report for Seoul on March 2023***\n")
    print("Line 1:")
    print("Busiest Station: {0:s} ({1:d})".format(bs1,bs1u))
    print("Least used Station: {0:s} ({1:d})\n".format(lbs1,lbs1u))
    print("Line 2:")
    print("Busiest Station: {0:s} ({1:d})".format(bs2,bs2u))
    print("Least used Station: {0:s} ({1:d})\n".format(lbs2,lbs2u))
    print("Line 3:")
    print("Busiest Station: {0:s} ({1:d})".format(bs3,bs3u))
    print("Least used Station: {0:s} ({1:d})\n".format(lbs3,lbs3u))
    print("Line 4:")
    print("Busiest Station: {0:s} ({1:d})".format(bs4,bs4u))
    print("Least used Station: {0:s} ({1:d})\n".format(lbs4,lbs4u))
    f.close()
    
if __name__ == '__main__':
    main() 
```
